```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv(f"/Users/ratirajpandav/Downloads/archive-3/Data.csv")

# Select relevant indicator (Change this to any other indicator if needed)
indicator = "Adolescent fertility rate (births per 1,000 women ages 15-19)" 

# Filter the data
df_filtered = df[df["Indicator.Name"] == indicator][["Country.Name", "2015"]].dropna()

# Sort values by highest rates
df_top10 = df_filtered.sort_values(by="2015", ascending=False).head(10)

# Plot
plt.figure(figsize=(12, 6))
sns.barplot(x=df_top10["2015"], y=df_top10["Country.Name"], palette="coolwarm")

plt.xlabel("Fertility Rate (Births per 1,000 Women)")
plt.ylabel("Country")
plt.title("Top 10 Countries by Adolescent Fertility Rate (2015)")
plt.show()
```
